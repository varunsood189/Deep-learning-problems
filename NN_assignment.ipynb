```python
features_test = pd.read_csv('test.csv')
features_test
```

```python
# Labels are the values we want to predict
labels_test = np.array(features_test['Winner (team 1=1, team 2=0)'])
# Remove the labels from the features
# axis 1 refers to the columns

for i,value in enumerate(features_test['Team 1']):
    if value.split(" ")[0]==features_test['CityOfGame'][i]:
        features_test['CityOfGame'][i]=1
    else:
        features_test['CityOfGame'][i]=2

features_test= features_test.drop('Team 1', axis = 1)
features_test= features_test.drop('Team 2', axis = 1)
features_test= features_test.drop('Game ID', axis = 1)
features_test= features_test.drop('Day', axis = 1)
#features_test= features_test.drop('CityOfGame', axis = 1)
features_test= features_test.drop('DateOfGame', axis = 1)
features_test= features_test.drop('TimeOfGame', axis = 1)
features_test= features_test.drop('Winner (team 1=1, team 2=0)', axis = 1)
# Saving feature names for later use
feature_list = list(features_test.columns)

features_test = np.array(features_test)
```

```python
t_label=np.zeros((labels_test.shape[0],2))
t_label.shape
test_features=features_test
test_labels=labels_test
```

```python
print('Training Features Shape:', train_features.shape)
print('Training Labels Shape:', train_labels.shape)
print('Testing Features Shape:', test_features.shape)
print('Testing Labels Shape:', test_labels.shape)
```

```python
#Normalizing the data
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
test_features = sc.fit_transform(test_features)
train_features=sc.fit_transform(train_features)
```

```python
print(train_features)
```

```python
#Dependencies
import keras
from keras.utils import plot_model
from keras.models import Sequential
from keras.layers import Dense, Dropout
# Neural network
model = Sequential()
model.add(Dense(16, input_dim=23, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
print(model.summary())
# plot graph
plot_model(model, to_file='multilayer_perceptron_graph.png')
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
X_train=train_features
y_train=train_labels
```

```python
history = model.fit(X_train, y_train, epochs=100, batch_size = 20)
```

```python
y_pred = model.predict(X_train)
#Converting predictions to label
pred = list()
for i in range(len(y_pred)):
    pred.append(np.argmax(y_pred[i]))
#Converting one hot encoded test label to label
# Use the forest's predict method on the test data
pred=[]
for i in y_pred:
    if i>0.5:
       pred+=[1]
    else:
       pred+=[0]
# Calculate the absolute errors
errors = abs(pred - labels)

print('Accuracy for the training set :', 1-round(np.mean(errors), 2))
```
